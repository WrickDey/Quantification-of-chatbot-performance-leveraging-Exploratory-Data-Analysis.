```python
dftemp=pd.DataFrame(columns=["Convno."])
for i in range(1369):
    k=0
    while (True):
        try:
            if df["turns"][i][k]["labels"]["acts"][0]["name"]!="affirm":
                k+=1
            elif df["turns"][i][k]["labels"]["acts"][0]["name"]=="affirm":
                break
        except Exception as e:
            k=-1
            break
        
    dftemp.loc[i,:]=k
   
```

```python
dftemp
```

```python
dftemp.value_counts()
```

```python
plt.figure(figsize=(8,5))
plt.hist(dftemp["Convno."],bins=[0,10,20,30,40,50,60,70],edgecolor="black")
plt.xlabel("No. of Conversations which led to booking confirmation")
plt.ylabel("Frequency of such conversations")
plt.axvline(dftemp.value_counts().median(),color="orange",label="Median no. of convos.")
plt.title("Histogram of Chatbot conversations that led to bookings")
plt.legend()
```

```python
dff=pd.read_csv("C:/Users/arnab/Desktop/.ipynb_checkpoints/EDA_file.csv")
```

```python
dff.head()
```

```python
dft=pd.DataFrame(columns=["Label","Average","Maximum","Minimum"])
dft
```

```python
dff.drop(columns=["Unnamed: 0"],inplace=True)
dff
```

```python
for i in range(1,380):
    dft.loc[i,"label"]=i
    k=dff[dff["label"]==i]
    dft.loc[i,"Average"]=np.mean(k.loc[:,"timestamp"])
    dft.loc[i,"Maximum"]=np.max(k.loc[:,"timestamp"])
    dft.loc[i,"Minimum"]=np.min(k.loc[:,"timestamp"])
        
        
```

```python
dft.drop(columns=["Label"],inplace=True)
```

```python
dft["label"]=dft["label"].astype(int)
```

```python
dft.set_index("label",inplace=True)
```
